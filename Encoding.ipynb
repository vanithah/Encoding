# label encoding

```python
# when we have categorical variable with ordinal values, then we can go with label encoding.
```

```python
data= pd.read_csv('Loan_data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['Property_Area'].value_counts()
```

```python
le = LabelEncoder()
data['Property_Area']=le.fit_transform(data['Property_Area'])
data.head(5)
```

```python
data['Property_Area'].value_counts()
```

```python
# Note: If label encoding is not assignng the values in order expected, then better go with replace() or map() 
```
